## Part 1: The Ideal Case

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# 1.

g = 9.81 # [m/s^2]

# dx = v0*t + 1/2 g*t^2

dx = 4e3 # m
t_analytic = np.sqrt( 2*dx / g )

print(f"The time to fall 4 km is t = {t_analytic} s.")
```

\#2.

The equation written as a coupled ystem of differential equations is

\begin{gather}
\frac{dy}{dt} = v \\
\frac{d v}{dt} = -g - \alpha v^{\gamma}
\end{gather}

```python
# 3.

# Define derivitives function
def ddt(t, s):
    """"s[0] = y
        s[1] = v = dydt"""
    global g, alpha, gamma
    
    y = s[0]
    v = s[1]

    D = np.zeros(2)
    D[0] = v # dy/dt
    D[1] = -g + alpha * v**gamma # d^2y/dt^2

    return D

    
# Solve Diff-eq

# Timespan
t0, tf = 0, 30 # [s]

n = 500
t_eval = np.linspace(t0,tf, n)

# ICs
ics = [0, 0] # y(0) = 0 m, v(0) = 0 m/s
alpha, gamma = 0, 0 # no drag

y_floor = -4000 # [m]

# Solve

# Event for when mass reaches the bottom
def bottom(t, s):
    # s[0] = y
    global y_floor
    return s[0] - y_floor
    
soln = solve_ivp(fun=ddt, t_span=(t0,tf), y0=ics, t_eval=t_eval, events=bottom)

# Plot Diff-eq
t, y, v = soln.t, soln.y[0], soln.y[1]

fig, ax = plt.subplots()
ax2 = ax.twinx()

ax.plot(t, y, c='g')
ax2.plot(t,v, c='r')

ax.hlines(y=y_floor, xmin=t0, xmax=tf, linestyles="--") # Plot floor

ax.set_xlabel("Time [s]")
ax.set_ylabel("Distance Travelled [m]")
ax.yaxis.label.set_color('green')

ax2.set_ylabel("Velocity [m/s]")
ax2.yaxis.label.set_color('red')

ax.set_title("Figure 1: Freefall Position and Velocity vs. Time")


# When the mass reaches the bottom
t_floor = soln.t_events[0][0]

print(f"The mass hits the floor at t = {t_floor} s.")
print(f"The difference between this calculate time and the time calculated analytically is {t_floor-t_analytic} s.\
This minute discrepancy is from a combination of rounding errors due to floating point arithmetic and error from \
the numerical method used to solve the differential equation.")




```

# Part 2: Including Drag and a Variable g

```python
# Define gravitional accelerate as a function of distance
def g(r):
    """ Returns gravitational acceleration as a function of r, valid for -R_E < r < 0
        Defined such that r = 0 is at the surface, r = -R_E is at the center of the Earth
    """
    R_E = 6378.1e3 # [m]
    g0 = 9.81 # [m/s^2]

    return g0 * ( (r + R_E) / R_E)


# Define derivitives function, now with g(y)
def ddt(t, s):
    """"s[0] = y
        s[1] = v = dydt"""
    global g, alpha, gamma
    
    y = s[0]
    v = s[1]

    D = np.zeros(2)
    D[0] = v # dy/dt
    D[1] = -g(y) + alpha * np.abs(v)**gamma # d^2y/dt^2

    return D

    
# Solve Diff-eq

# Timespan
t0, tf = 0, 30 # [s]

n = 500
t_eval = np.linspace(t0,tf, n)

# ICs
ics = [0, 0] # y(0) = 0 m, v(0) = 0 m/s
alpha, gamma = 0, 0 # no drag

y_floor = -4000 # [m]


# Solve

# Event for when mass reaches the bottom
def bottom(t, s):
    # s[0] = y
    global y_floor
    return s[0] - y_floor
    
soln = solve_ivp(fun=ddt, t_span=(t0,tf), y0=ics, t_eval=t_eval, events=bottom)

# Plot Diff-eq
t, y, v = soln.t, soln.y[0], soln.y[1]

fig, ax = plt.subplots()
ax2 = ax.twinx()

ax.plot(t, y, c='g')
ax2.plot(t,v, c='r')

ax.hlines(y=y_floor, xmin=t0, xmax=tf, linestyles="--") # Plot floor

ax.set_xlabel("Time [s]")
ax.set_ylabel("Distance Travelled [m]")
ax.yaxis.label.set_color('green')

ax2.set_ylabel("Velocity [m/s]")
ax2.yaxis.label.set_color('red')

ax.set_title("Figure 2: Freefall with variable g(y)")

# When the mass reaches the bottom
t_floor2 = soln.t_events[0][0]


# 2.
print(f"The new fall time, now with variable acceleration, is {t_floor2} s.")
print(f"Considering a height-dependent g will increase the time it takes \
for a mass to fall, because, as the mass falls, the proportion of the Earth's mass interior to the test mass decreases, \
so the force, and therefore the acceleration, will decrease. A decrease in acceleration on the mass will correspond to \
a greater freefall time. Quantitatively, the freefall time increases by {t_floor2 - t_floor} s.")
```

```python
# 3.

# Define new time span
t0, tf = 0, 100 # [s]

n = 500
t_eval = np.linspace(t0,tf, n)

# Define new ICs
gamma = 2
alpha = 0.004

soln = solve_ivp(fun=ddt, t_span=(t0,tf), y0=ics, t_eval=t_eval, events=bottom)

# Plot Diff-eq
t, y, v = soln.t, soln.y[0], soln.y[1]

fig, ax = plt.subplots()
ax2 = ax.twinx()

ax.plot(t, y, c='g')
ax2.plot(t,v, c='r')

ax.hlines(y=y_floor, xmin=t0, xmax=tf, linestyles="--") # Plot floor

ax.set_xlabel("Time [s]")
ax.set_ylabel("Distance Travelled [m]")
ax.yaxis.label.set_color('green')

ax2.set_ylabel("Velocity [m/s]")
ax2.yaxis.label.set_color('red')

ax.set_title("Figure 2.2: Freefall with variable g(y) and Drag")

print("To calibrate alpha, I tweaked its value until the velocity flattens out around v = -50 m/s.")


# Return Fall Time
t_floor3 = soln.t_events[0][0]
print(f"The fall time with drag is {t_floor3} s. Considering drag has a significant effect on the fall time, increasing \
it by {t_floor3 - t_floor2} s.")
```

# Part 3: Coriolis Force

```python
# Define derivitives function, now with a Coriolis term and in 2 dimensions x and y
def ddt(t, s):
    """"s[0] = x, s[1] = y
        s[2] = vx = dxdt, s[3] = vy = dy/dt
        s = [x,y vx,vy]
        y is the vertical coordinate, x is the transverse coordinate"""
    global g, alpha, gamma, omega
    
    x,y = s[0], s[1]
    vx,vy = s[2], s[3]

    D = np.zeros(4)
    
    D[0], D[1] = vx, vy # x,y 1st derivitives

    dxdt2 = -2 * omega * vy # d^2x/dt^2
    dydt2 = -g(y) + ( alpha * np.abs(vy)**gamma ) + (2 * omega * vx) # d^2y/dt^2
    
    D[2], D[3] = dxdt2, dydt2 # x,y 2nd derivitives

    return D


# Solve Diff-eq

# Timespan
t0, tf = 0, 100 # [s]

n = 500
t_eval = np.linspace(t0,tf, n)

# ICs
ics = [0,0, 0,0] # x(0) = y(0) = 0 m, vx(0) = vy(0) = 0 m/s
gamma = 2
alpha = 0.00 # no drag
omega = 7.272e-5 # [rad/s]

y_floor = -4000 # [m]


# Solve

# Event for when mass reaches the bottom
def bottom(t, s):
    # s[1] = y
    global y_floor
    return s[1] - y_floor

def wall(t0, s):
    # s[0] = x
    return np.abs(s[0]) - 2.5


bottom.terminal = True
    
soln = solve_ivp(fun=ddt, t_span=(t0,tf), y0=ics, t_eval=t_eval, events=(bottom, wall))

# Plot
x = soln.y[0]
y = soln.y[1]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel("Transerve Displacement [m]")
ax.set_ylabel("Depth [m]")
ax.set_title("Figure 3: Transverse vs Vertical Displacement with the Coriolis Force")

ax.hlines(y=y_floor, xmin=0, xmax=10, linestyles="--"); # Plot floor

# Plot dots every 20 time-steps (4 seconds)
xdots = []
ydots = []
for i in range(len(soln.t)):
    if i%20 == 0:
        xdots.append(x[i])
        ydots.append(y[i])

ax.scatter(xdots, ydots)

print(f"The mass hits the bottom at {soln.t_events[0][0]} s and hits the wall at {soln.t_events[1][0]} s, so it \
will hit the wall before reaching the bottom.")
```

```python
# 4. Now with drag
alpha = 0.004

soln = solve_ivp(fun=ddt, t_span=(t0,tf), y0=ics, t_eval=t_eval, events=(bottom, wall))

# Plot
x = soln.y[0]
y = soln.y[1]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel("Transerve Displacement [m]")
ax.set_ylabel("Depth [m]")
ax.set_title("Transverse vs Vertical Displacement with the Coriolis Force and Drag")

ax.hlines(y=y_floor, xmin=0, xmax=30, linestyles="--"); # Plot floor

# Plot dots every 20 time-steps (4 seconds)
xdots = []
ydots = []
for i in range(len(soln.t)):
    if i%20 == 0:
        xdots.append(x[i])
        ydots.append(y[i])

ax.scatter(xdots, ydots)


print(f"The mass hits the bottom at {soln.t_events[0][0]} s and hits the wall at {soln.t_events[1][0]} s, so it \
will still hit the wall before reaching the bottom. The drag makes a significant difference here, since the mass \
falls at a much lower rate, with the velocity maxing out at about 50 m/s.")
print(f"The mass hits the wall at a depth of {soln.y_events[1][0][1]} m.")
```

# Part 4: An Infinitely Deep Mine

```python
# Define gravitional accelerate as a function of distance
R_E = 6378.1e3 # [m]
def g(r):
    """ Returns gravitational acceleration as a function of r, valid for |r| <= R_E
        Defined such that r = 0 is at the center, r = R_E is at the surface of the Earth
    """
    global R_E
    g0 = 9.81 # [m/s^2]

    return g0 * ( r / R_E)

# Define derivitives function, now with g(y)
def ddt(t, s):
    """"s[0] = y
        s[1] = v = dydt"""
    global alpha, gamma
    
    y = s[0]
    v = s[1]

    D = np.zeros(2)
    D[0] = v # dy/dt
    D[1] = -g(y) + alpha * np.abs(v)**gamma # d^2y/dt^2

    return D


# Define events
def center(t, s):
    """s[0] = y
    when the mass reaches the center"""
    return s[0]

def otherside(t, s):
    """s[0] = y 
    when the mass reaches the other side i.e. y = -R_E"""
    global R_E
    
    return s[0] + 0.999*R_E

center.terminal = False
otherside.terminal = False
# Solve Diff-eq

# Timespan
t0, tf = 0, 5000 # [s]

n = 10000
t_eval = np.linspace(t0,tf, n)

# ICs
ics = [R_E, 0.] # y(0) = R_E m, v(0) = 0 m/s
alpha, gamma = 0, 0

# Solve
soln = solve_ivp(fun=ddt, t_span=(t0,tf), y0=ics, t_eval=t_eval, events=(center, otherside), rtol=1e-9)

# Plot Diff-eq
t, y, v = soln.t, soln.y[0], soln.y[1]

fig, ax = plt.subplots()
ax2 = ax.twinx()

ax.plot(t, y, c='g')
ax2.plot(t,v, c='r')

ax.set_xlabel("Time [s]")
ax.set_ylabel("Distance Travelled [m]")
ax.yaxis.label.set_color('green')

ax2.set_ylabel("Velocity [m/s]")
ax2.yaxis.label.set_color('red')

ax.set_title("Figure 4: Freefall down an Infinitely Deep Mine")

print("The plot shows that the position oscialltes in time about the center of the Earth, as the mass is always being pulled toward the center, \
and loses no energy.")

# 2.
t_otherside = soln.t_events[1][0]
print(f"The mass reaches the other side of the Earth at t = {t_otherside} s. \n\
The mass first reaches the center at t = {soln.t_events[0][0]} s with a velocity of v = {soln.y_events[0][0][1]} m/s.")

M_Earth = 5.972e24 # [kg]
G = 6.672e-11
T_orbit = np.sqrt( ( ( 4 * np.pi**2 ) / (G*M_Earth) )*R_E**3 ) # Using Keplers's 3rd Law

print(f"\nThe orbital period of a mass orbiting Earth at a distance of 1 Earth radius is {T_orbit} s. So, the time to fall through the mine is \
{T_orbit - t_otherside} s faster than the time to orbit, or about half the orbital period. This intuitively makes sense, because the path that \
the mass follows during an orbit is longer than the path it takes through the Earth, by a factor of pi. Though, the forces, and therefore the accelerations, \
the mass will feel are different, so the difference in path length does not tell the whole story")
```

# Part 5: A Non-Uniform Earth

```python
from scipy.integrate import quad
R_E = 6378.1e3 # [m]
M_E = 5.972e24 # [kg]


# Define density function
def rho(r, n):
    global R_E

    p = ( 1 - (r / R_E)**2 )**n
    return p

# Compute pn for n = 0,1,2,9
def integrand(r, n):
    """rho(r)dV in spherical symmetry"""
    return rho(r, n) * r**2
    
def norm_rho(n):
    """density normalization constant"""
    global R_E, M_E
    rho_n = M_E / ( 4 * np.pi * quad(integrand, 0, R_E, args=(n))[0] ) # Neglects uncertainty from quad
    return rho_n

pn = [norm_rho(0), norm_rho(1), norm_rho(2),0,0,0,0,0,0, norm_rho(9)] # For ease of reference


# 1. Plot Normalized Density Profile
n = 1000
r = np.linspace(0.1, R_E, n)

fig, ax = plt.subplots()
ax.plot(r, rho(r,0), label="n=0")
ax.plot(r, rho(r,1), label="n=1")
ax.plot(r, rho(r,2), label="n=2")
ax.plot(r, rho(r,9), label="n=9")

ax.set_xlabel("Radius [m]")
ax.set_ylabel("Density [kg/m^3]")
ax.set_title("Figure 5: Normalized Density Profiles")

plt.legend();
```

```python
# 2. Plot Force Profile

# Define Mass Enclosed Function
def M(r, n):
    """Returns the mass enclosed in a sphere of radius r with density profile rho"""
    
    M = 4*np.pi * pn[n] * quad(integrand, 0, r, args=(n))[0]
    return M
M = np.vectorize(M)

# Define Force Magnitude Function
def Force(r, n):
    """Returns the magnitude of the force vector using Newton's Law of Gravitation"""
    G = 6.674e-11 # [N m^2 / kg^2]
    
    if np.isclose(r,0) == True: # Remove singularity (at r=0, M=0 so F=0)
        return 0
    else:
        F = ( G * M(r, n) ) / r**2
    return F
Force = np.vectorize(Force)


# Plot
fig, ax = plt.subplots()
ax.plot(r, Force(r,0), label="n=0")
ax.plot(r, Force(r,1), label="n=1")
ax.plot(r, Force(r,2), label="n=2")
ax.plot(r, Force(r,9), label="n=9")

ax.set_xlabel("Radius [m]")
ax.set_ylabel("Gravitational Force [N]")
ax.set_title("Figure 6: Force Profiles")

plt.legend();
```

```python
# 3. Plot Position and Velocity

Force = np.vectorize(Force)

# Define derivitives functions
def deriv(t, s):
    """ s[0] = y, s[1] = v"""
    global n
    y = s[0]
    v = s[1]

    dydt = v
    dvdt = -Force(y, n) # a = F/m, m = 1

    return [dydt, dvdt]

# Define event function
def center(t, s):
    """s[0] = y
    when the mass reaches the center"""
    return s[0]


# Timespan
t0, tf = 0, 4000 # [s]

N = 5000
t_eval = np.linspace(t0,tf, N)

# Solve
n = 0
soln0 = solve_ivp(fun=deriv, t_span=(t0,tf), y0=ics, t_eval=t_eval, events=center)

n = 1
soln1 = solve_ivp(fun=deriv, t_span=(t0,tf), y0=ics, t_eval=t_eval, events=center)

n = 2
soln2 = solve_ivp(fun=deriv, t_span=(t0,tf), y0=ics, t_eval=t_eval, events=center)

n = 9
soln9 = solve_ivp(fun=deriv, t_span=(t0,tf), y0=ics, t_eval=t_eval, events=center)
```

```python
# Plot
fig, axs = plt.subplots(2,2, figsize=(12,8))
ax0, ax1, ax2, ax9 = axs[0][0], axs[0][1], axs[1][0], axs[1][1]

ax0v, ax1v, ax2v, ax9v = ax0.twinx(), ax1.twinx(), ax2.twinx(), ax9.twinx()

def plot(ax_y, ax_v, soln, nlabel):
    """function to plot solution results"""
    y, v, t = soln.y[0], soln.y[1], soln.t
    
    ax_y.plot(t, y, c='g')
    ax_v.plot(t, v, c='r')
    
    ax_y.set_xlabel("Time [s]")
    ax_y.set_ylabel("Distance from Center [m]")
    ax_y.yaxis.label.set_color('green')
    
    ax_v.set_ylabel("Velocity [m/s]")
    ax_v.yaxis.label.set_color('red')

    ax_y.set_title(f"n = {nlabel}")

plot(ax0, ax0v, soln0, 0)
plot(ax1, ax1v, soln1, 1)
plot(ax2, ax2v, soln2, 2)
plot(ax9, ax9v, soln9, 9)
fig.suptitle("Figure 7: Position and Velocity through a Non-Uniform Earth")
plt.tight_layout()

# Time and Speed at Center
# n = 0
t_center0 = soln0.t_events[0][0]
v_center0 = soln0.y_events[0][0][1]

# n = 1
t_center1 = soln1.t_events[0][0]
v_center1 = soln1.y_events[0][0][1]

# n = 2
t_center2 = soln2.t_events[0][0]
v_center2 = soln2.y_events[0][0][1]

# n = 9
t_center9 = soln9.t_events[0][0]
v_center9 = soln9.y_events[0][0][1]

print(f"For n = 0, the time to reach the center of is t = {t_center0} s, with a velocity at that time of v = {v_center0} m/s.")
print(f"For n = 1, the time to reach the center of is t = {t_center1} s, with a velocity at that time of v = {v_center1} m/s.")
print(f"For n = 2, the time to reach the center of is t = {t_center2} s, with a velocity at that time of v = {v_center2} m/s.")
print(f"For n = 9, the time to reach the center of is t = {t_center9} s, with a velocity at that time of v = {v_center9} m/s.")
```

# Part 9: A Lunar Mineshaft

```python
M_M = 7.35e22 # [kg] Lunar mass
R_M = 1738.1e3 # [m]  Lunar radius

# Redefine functions using Moon parameters
# Define density function
def rho(r, n):
    global R_M

    p = ( 1 - (r / R_M)**2 )**n
    return p
    
def norm_rho(n):
    """density normalization constant"""
    global R_M, M_M
    rho_n = M_M / ( 4 * np.pi * quad(integrand, 0, R_M, args=(n))[0] ) # Neglects uncertainty from quad
    return rho_n

# Define Mass Enclosed Function
def M(r, n):
    """Returns the mass enclosed in a sphere of radius r with density profile rho"""
    
    M = 4*np.pi * norm_rho(n) * quad(integrand, 0, r, args=(n))[0]
    return M
M = np.vectorize(M)

# Time span
t0, tf = 0, 4000
t_eval = np.linspace(t0,tf, 2000)

# ICs
n = 0 # Assume constant density
ics = [R_M, 0] # y = R_M, v = 0 m/s

# Solve
soln = solve_ivp(fun=deriv, t_span=(t0,tf), y0=ics, t_eval=t_eval, events=center)

t_center_M = soln.t_events[0][0]
print(f"The time to fall to the center of the Moon, assuming constant density, is t = {t_center_M} s.")


# Densities
rho_E = M_E / ( (4/3)*np.pi*R_E**3 )
rho_M = M_M / ( (4/3)*np.pi*R_M**3 )
print(f"The average density of the Earth is {rho_E} kg/m^3. The average density of the moon is {rho_M} kg/m^3.")


# find density relationship
def fall_time():
    # Time span
    t0, tf = 0, 4000
    t_eval = np.linspace(t0,tf, 2000)
    
    # ICs
    n = 0 # Assume constant density
    ics = [R_M, 0] # y = R_M, v = 0 m/s
    
    # Solve
    soln = solve_ivp(fun=deriv, t_span=(t0,tf), y0=ics, t_eval=t_eval, events=center)
    
    t_fall = soln.t_events[0][0]

    return t_fall

"""density is directly proportional to mass, so finding a relationship between fall time and mass finds the relationship between fall time and density"""

# Refine density normalization function with new mass
def norm_rho(n):
    """density normalization constant"""
    global R_M, M_M
    rho_n = 4*M_M / ( 4 * np.pi * quad(integrand, 0, R_M, args=(n))[0] ) # Neglects uncertainty from quad
    return rho_n

t_fall_4M = fall_time()

def norm_rho(n):
    """density normalization constant"""
    global R_M, M_M
    rho_n = 16*M_M / ( 4 * np.pi * quad(integrand, 0, R_M, args=(n))[0] ) # Neglects uncertainty from quad
    return rho_n

t_fall_16M = fall_time()

```

```python
times = [t_center_M, t_fall_4M, t_fall_16M]
dens = [rho_M, 4*rho_M, 16*rho_M]

# Plot fall_time vs density to find the relationship
fig, ax = plt.subplots()
ax.scatter(dens, times, label="Computed Times")
ax.set_xlabel("Density [kg/m^3]")
ax.set_ylabel("Fall Time [s]")
ax.set_title("Crossing Time vs. Density")

pp = np.linspace(rho_M, 16*rho_M, 10000)

# Fit to exponential
def fit(x, A, n):
    return A * x**n

from scipy.optimize import curve_fit

fit_par, pcov = curve_fit(fit, xdata=dens, ydata=times)
print(f"The power of the exponential is {fit_par[1]}.")

ax.plot(pp, fit(pp, fit_par[0], fit_par[1]), label="1/sqrt(x) Fit")
plt.legend();

print("The fall time seems to be proportional to density^(-1/2).")
```
