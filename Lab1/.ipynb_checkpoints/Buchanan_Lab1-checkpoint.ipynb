## Part 1: The Gravitational Potential of the Earth

```python
# 1.
import numpy as np
import matplotlib.pyplot as plt

# Gravitational Potential Function
def gravPot(xm, ym, x, y, M):
    """
    Returns the gravitational potential at a point (x,y)
    due to a mass M located at the point (xm,ym)
    """
    G = 6.67e-11 # [m^3 s^2 kg^-1 ]

    # Compute the distance r from the evaluation point to the mass location
    r = np.sqrt( (x - xm)**2 + (y - ym)**2 )

    # Compute the potential
    phi = -(G * M) / r

    return phi
```

```python
# 2.

n = 5000
# x-vals from 1 Earth-Radius to 1.5 times the distance to the Moon
x = np.linspace(6378e3, 1.5*3.8e8, n) # [m]
y = np.zeros(n) # [m]

# Set Earth parameters

# Location of Earth (set to origin)
xm = 0 # [m]
ym = 0 # [m]
M_Earth = 5.9e24 # [kg]

# Evaluate Function
pots = np.abs(gravPot(xm, ym, x, y, M_Earth))


# Plot
fig, ax = plt.subplots()
ax.plot(x, pots)
ax.set_yscale('log')
ax.set_xlabel('Distance [m]')
ax.set_ylabel('Gravitational Potential [J/kg]')
ax.set_title("Gravitational Potential of Earth vs Distance")
```

```python
# 3.
import matplotlib
import matplotlib.cm as cm

xm = 0 # [m]
ym = 0 # [m]

n3 = 2000
xvals = np.linspace(-1.5*3.8e8, 1.5*3.8e8, n3) # [m]
yvals = np.linspace(-1.5*3.8e8, 1.5*3.8e8, n3) # [m]

x_grid, y_grid = np.meshgrid(xvals, yvals)

# Evaluate Potential
pots2 = np.abs(gravPot(xm, ym, x_grid, y_grid, M_Earth))

# Plot
fig2, ax2 = plt.subplots()
mesh = ax2.pcolormesh(x_grid, y_grid, pots2, norm=matplotlib.colors.LogNorm(), cmap=cm.hot)
ax2.set_aspect('equal') 
ax2.set_xlabel('x [m]') 
ax2.set_ylabel('y [m]') 
ax2.set_title('Earth\'s Gravitational Potential Field')

cbar = fig2.colorbar(mesh)
cbar.set_label('Gravitational Potential [J/kg]')
```

## Part 2: Gravitational Potential of the Earth-Moon System

```python
# 1.

# Set x and y values
n = 2000
xvals = np.linspace(-1.5*3.8e8, 1.5*3.8e8, n) # [m]
yvals = np.linspace(-1.5*3.8e8, 1.5*3.8e8, n) # [m]

x_grid, y_grid = np.meshgrid(xvals, yvals)

# Evaluate Earth's Potential
(xE, yE) = (0,0) # [m]
M_Earth = 5.9e24 # [kg]
E_pot = np.abs(gravPot(xE, yE, x_grid, y_grid, M_Earth)) # [J/kg]

# Evaluate Moon's Potential
(xM, yM) = ( (1.5*3.8e8)/np.sqrt(2), (1.5*3.8e8)/np.sqrt(2) ) # [m]
M_Moon = 7.3e22 # [kg]
M_pot = np.abs(gravPot(xM, yM, x_grid, y_grid, M_Moon)) # [J/kg]

# Compute Total Potential as the Superposition of the Earth and Moons Potentials
T_pot = E_pot + M_pot # [J/kg]

# Plot
fig, ax = plt.subplots()
mesh = ax.pcolormesh(x_grid, y_grid, T_pot, norm=matplotlib.colors.LogNorm(), cmap=cm.hot)
ax.set_aspect('equal') 
ax.set_xlabel('x [m]') 
ax.set_ylabel('y [m]') 
ax.set_title('Gravitational Potential Field of Earth-Moon System')

cbar = fig.colorbar(mesh)
cbar.set_label('Gravitational Potential [J/kg]')
```

```python
# 2.

# Plot Contour using the same values as 1.
fig2, ax2 = plt.subplots()

contour = ax2.contour(x_grid, y_grid, T_pot, levels=np.linspace(0,1e8, 175), linewidths=0.75, norm=matplotlib.colors.LogNorm(),  cmap=cm.hot)
ax2.set_xlabel('x [m]')
ax2.set_ylabel('y [m]')
ax2.set_aspect('equal')
ax2.set_title("Contours of Earth-Moon Gravitational Potential");
```

## Part 3: The Gravitational Force Field of the Earth-Moon System

```python
# 1.

def gravForce(xM, yM, xm, ym, M, m):
    G = 6.674e-11 # [m^3 s^2 kg^-1]

    # Define displacement vector
    r_vec = np.array( [xm - xM, ym - yM] )
    r_hat = r_vec / np.linalg.norm(r_vec) # r unit vector
    r_mag_sq = (xm - xM)**2 + (ym - yM)**2 # r square magnitude

    # Compute Force vector
    F_mag = (G * M*m) / r_mag_sq # Force magnitude
    F_vec = -F_mag * r_hat # Force vector

    Fx = F_vec[0]
    Fy = F_vec[1]

    return Fx, Fy
```

```python
# 2.
# Define Parameters (Distances in meters)
M_Earth = 5.9e24 # [kg]
M_Moon = 7.3e22 # [kg]
(xE, yE) = (0, 0) # Earth Position [m]
(xM, yM) = ( (3.8e8)/np.sqrt(2), (3.8e8)/np.sqrt(2) ) # Moon Position [m]

m = 5500 # Apollo 11 Command Module Mass [kg]

# Set x and y values
n = 4000
xvals = np.linspace(-1.5*3.8e8, 1.5*3.8e8, n) # [m]
yvals = np.linspace(-1.5*3.8e8, 1.5*3.8e8, n) # [m]

x_grid, y_grid = np.meshgrid(xvals, yvals)

# Compute Earth Field
Fx_E, Fy_E = gravForce(xE, yE, x_grid, y_grid, M_Earth, m)

# Compute Moon Field
Fx_M, Fy_M = gravForce(xM, yM, x_grid, y_grid, M_Moon, m)

# Compute Total Field
Fx = Fx_E + Fx_M
Fy = Fy_E + Fy_M
F_mag = np.sqrt( Fx**2 + Fy**2 )



# Plot
import matplotlib.cm as cm

fig, ax = plt.subplots()

colors = F_mag
stream = ax.streamplot(x_grid, y_grid, Fx, Fy, color=colors, norm=matplotlib.colors.LogNorm(), cmap=cm.plasma, density=2)
cbar = fig.colorbar(stream.lines)
cbar.set_label("Force Magnitude [N]")
ax.set_xlabel("x [m]")
ax.set_ylabel("y [m]")
ax.set_title("Gravitational Force Field of the Earth-Moon System")

ax.plot(xM,yM, marker='o', color='grey') # Plot Moon
ax.plot(xE,yE, marker='o', color='green'); # Plot Earth
```

## Part 4: Altitude of the Saturn V Rocket

```python
# 1.

# Define parameters
m0 = 2.8e6 # Wet Mass [kg]
mf = 7.5e5 # Dry Mass [kg]
mdot = 1.3e4 # Burn rate [kg/s]
v_e = 2.4e3 # Exhaust velocity [m/s]
g = 9.81 # [m/s^2]

# Calculate Burn Time
T = (m0 - mf) / mdot # s


# 2.
# Define Change in Velocity Function
def delVel(t, m0, mf, mdot, v_e, g=9.81):

    if m0 - mdot*t > mf:
        delv = v_e * np.log( m0 / (m0 - mdot*t) ) - g*t
    else:
        delv = 0

    return delv

dVel = np.vectorize(delVel)

# Define time range
t = np.linspace(0, T, 500)

# Compute deltaV over time
dv = dVel(t, m0, mf, mdot, v_e) # [m/s]

fig, ax = plt.subplots()
ax.plot(t,dv);
ax.set_xlabel('Time [s]')
ax.set_ylabel('Change in Velocity [m/s]')
ax.set_title("Change in Velocity vs Time")


# 4.
from scipy.integrate import quad

# Compute the altitude
h, herr = quad(dVel, 0, T, args=(m0, mf, mdot, v_e))

print(f"The altitude after the burnout time T = {T:.2f} s is {h:.2f} m.")
```
